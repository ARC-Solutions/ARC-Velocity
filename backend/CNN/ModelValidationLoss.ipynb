```python
# Use a different style, 'ggplot' is quite modern looking, but there are others like 'seaborn'
plt.style.use('ggplot')

# You can still use dark background if you like it
# plt.rcParams['axes.facecolor'] = 'black'
plt.style.use('dark_background')
print(plt.style.available)

# Increase the size of the plot
fig, ax = plt.subplots(figsize=plot_size)

# Set title parameters
plt.suptitle(plot_title, fontproperties=title_font, size=title_font_size, color='white')

# Plot data with a modern look
plt.plot(history.history['loss'], marker='o', markersize=5, linewidth=2, color='c', label='Training loss')
plt.plot(history.history['val_loss'], marker='o', markersize=5, linewidth=2, color='red', label='Validation loss')

# Set labels with larger font size
plt.ylabel('Loss', fontsize = axis_label_font_size, fontproperties=label_font, color='white')
plt.xlabel('Epoch', fontsize = axis_label_font_size, fontproperties=label_font, color='white')

# Increase the size of tick labels and change their color
plt.xticks(fontsize=tick_label_font_size, color='white', fontproperties=label_font)
plt.yticks(fontsize=tick_label_font_size, color='white', fontproperties=label_font)

# Enable legend and grid
legend = plt.legend(loc='best', prop=label_font)
for text in legend.get_texts():
    text.set_color('white')
plt.grid(True, linestyle='--', alpha=0.6)

add_logo()
plt.tight_layout()
plt.show()
```
